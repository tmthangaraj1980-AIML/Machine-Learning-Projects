## STEP 3 — Advanced Feature Engineering

####What is Feature Engineering?

Feature engineering means:

* creating new intelligent columns from existing data.

* Machine Learning models learn better when we provide meaningful features.

###Goal of This Step

We will create:

* customer behavior features
* shipment risk features
* logistics intelligence features
* business logic features

These improve prediction accuracy.

###STEP 3.1 — Create Copy of Dataset

```python
# Create copy of dataset

df_fe = df.copy()
```

###STEP 3.2 — Discount Category Feature
Business Logic

Large discounts may increase:

* order volume
* delivery pressure
* delays

```python
# Create Discount Category

df_fe['Discount_Level'] = pd.cut(
    df_fe['Discount_offered'],
    bins=[0,10,20,40,100],
    labels=['Low','Medium','High','Very_High']
)

df_fe[['Discount_offered','Discount_Level']].head()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Discount_Level',
    hue='Reached.on.Time_Y.N',
    data=df_fe
)

plt.title('Discount Level vs Delivery Status')

plt.legend(['On-Time','Delayed'])

plt.show()
```

###STEP 3.3a — Complaint ratio

Complaint_Ratio = risk-related feature


```python
# Complaint Ratio
df_fe['Complaint_Ratio'] = (
    df_fe['Customer_care_calls'] /
    (df_fe['Prior_purchases'] + 1)
)
```

### step 3.3b - Loyal_Customer = loyalty/customer behavior feature

```python
# Loyal Customer
df_fe['Loyal_Customer'] = np.where(
    df_fe['Prior_purchases'] > 3,
    1,
    0
)
```

###STEP 3.5 — Shipping Efficiency Score
Business Logic

Higher customer rating with fewer support calls indicates:

* smoother delivery experience

```python
# Shipping Efficiency Score

df_fe['Shipping_Efficiency'] = (
    df_fe['Customer_rating'] /
    (df_fe['Customer_care_calls'] + 1)
)

df_fe[['Shipping_Efficiency']].head()
```

###STEP 3.6 — Cost Per Weight Feature
Business Logic

This helps understand:

* expensive lightweight products
* cheap heavy products

```python
# Cost Per Weight

df_fe['Cost_Per_Weight'] = (
    df_fe['Cost_of_the_Product'] /
    df_fe['Weight_in_gms']
)

df_fe[['Cost_Per_Weight']].head()
```

## step 4. Data preprocessing

```python
df_fe.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
categorical_cols = ['Warehouse_block',
           'Mode_of_Shipment',
           'Product_importance',
           'Gender',
           'Discount_Level']
for col in categorical_cols:
    df_fe[col] = le.fit_transform(df_fe[col])
```

```python
df_fe.info()
```

### step 4.1 Separate Features and Target

```python
X = df_fe.drop(columns = ['ID', 'Reached.on.Time_Y.N'])
y = df_fe['Reached.on.Time_Y.N']
```

### step 4.2 Train_Test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,
              stratify= y, random_state = 12)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### step 4.3 Feature scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Step 5. Model Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

## Step 6 Model Evaluation

```python
y_pred =model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
