```python
# Drop ID column
df = df.drop('CASENUM', axis=1)

# Target
y = df['ATTORNEY']

# Features
X = df.drop('ATTORNEY', axis=1)
```

```python
#train_test split
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=32, shuffle=True)
```

```python
X_train
```

```python
y_train
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
accuracy_train = accuracy_score(y_train, y_train_pred)
print('Accuracy:', accuracy_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_test = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(precision_score(y_test, y_pred))
```

```python
print(recall_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

##Precision = out of total positives predicted how  many actually positive? - How often we right?TP/(TP+FP)
##Recall = "Of all the actual positive cases, how many did the model correctly identify?" TP/(TP+FN)
