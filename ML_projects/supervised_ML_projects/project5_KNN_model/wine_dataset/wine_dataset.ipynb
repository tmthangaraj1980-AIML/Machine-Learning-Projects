```python
import pandas as pd
from sklearn.datasets import load_wine
from
```

```python
df = load_wine()
df = pd.DataFrame(data=df.data, columns=df.feature_names)
df
```

```python
df['target'] = load_wine().target
```

```python
X = df.drop(['target'], axis = 1)
y =df['target']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st = StandardScaler()
scaled_X = st.fit_transform(X)
```

```python
scaled_X
```

```python
X=pd.DataFrame(data=scaled_X, columns=X.columns)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify=y, random_state=42, shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train, y_train)
```

```python
y_train_pred = knn.predict(X_train)
```

```python
print("Train_accuracy", accuracy_score(y_train, y_train_pred))
```

```python
y_pred = knn.predict(X_test)
```

```python
print("Test_accuracy", accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
cv = []
for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn_fold = KFold(n_splits=5, shuffle = True, random_state=42)
    cross_val = cross_val_score(knn, X_train, y_train, cv=knn_fold)
    cv.append(cross_val.mean())

```

```python
max(cv)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,30), cv)
```

```python
cv.index(max(cv))
```
